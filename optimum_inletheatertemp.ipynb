## A Cost Based Model For Clean Cycle Optimization

A heat exchanger network is a combination of several heat exchangers
inter­connected. <br> Figure 7.3 shows a typical example of the heat exchanger
network where feed is preheated by several process streams before being
heated by a furnace to a desirable temperature required for
fractionation.<br> The terminal temperature of the exchanger network
($T^{N}$) is the same as the furnace inlet temperature (FIT). <br> Clearly,
the network terminal temperature represents collective heat recovery by
individual heat exchangers in the network.

Thus, the overall performance for a heat exchanger network can be
measured by the network terminal temperature ($T^{N}$). $T^{N}$ declines
over time due to exchanger fouling resulting in increased furnace duty.<br>
![Example Image](imgs/heat_train1.png) <br>
To deﬁne a nominal base case to focus on fouling alone, typical feed
rate, standard physical properties, and temperatures for all streams
entering the network are used to model the heat exchanger network. The
terminal temperature calculated for the nominal case is called nominal
terminal temperature. To calculate the nominal terminal temperature at
any point of time, the heat transfer coefﬁcients for heat exchangers are
calculated using spot online readings of temperature and ﬂow rates based
on equation (6.1) U<sub>i</sub>A<sub>i</sub>
=Q<sub>i</sub>/△T<sub>LTMD,i</sub>. The UA values calculated are then
used in the heat exchanger network based on the nominal case to
determine the network terminal temperature.

Let us assume the plant starts operation with a clean set of exchangers.
After t<sub>1</sub> days of operation, the exchangers appear to be much
fouled and the plant engineers may debate if they should clean the
exchangers to restore the network terminal temperature. Assume the
process unit is shut down as t2 days are required for cleaning. On one
side of the dilemma, fouling increases the furnace duty and thus
operating cost. On the other hand, cleaning will take t<sub>2</sub> days
at lost margin and K dollars as cleaning costs. Therefore, does it make
sense to do the cleaning now or should it be postponed? The quest is to
search for the optimal cleaning cycle in terms of time interval
t<sub>1</sub> with cleaning time t<sub>2</sub> known in advance.

It has been observed by Smeenk and Kers (2001) that the network terminal
temperature declines linearly with time due to fouling. If denoting
($T^{N}$) for the nominal network terminal temperature at any time and
($T^{N}$) for the nominal network terminal temperature at the clean
condition, respectively, the linear decline of ($T^{N}$) with time can be
expressed as

$$T^{N}(t) = T_{0}^{N} - \lambda.t$$

where t stands for time (e.g., days) and lambda is the decay of ($T^{N}$) due
to fouling in °C/ day. The SI units are used by Smeenk and Kers (2001)
in deriving the equations below.

Similarly, the fact that furnace duty Q increases linearly as
T<sup>N</sup> decreases over time can be modeled via

$$Q(t) = Q_{0}^{N}\  - \ \beta.T^{N}(t)$$

where beta is the rate of heater duty reduction as $T^{N}$ declines over
time and  $$Q_{0}^{N}$$
is the heater duty at initial time of data points for generating the
correlation in equation (7.4).

Thus, the furnace fuel cost can be expressed as

$$E(t)\  = c_{f}Q(t)\  = \ c_{f}\left\lbrack Q_{0}^{N}\  - \ \beta.T^{N}(t) \right\rbrack\  = \ A\  - \ B\ .T^{N}(t)$$

where c<sub>f</sub> is the fuel price in \$/GJ, A = c<sub>f</sub> .
$Q_{0}^{N}$ , and B = c<sub>f</sub> . $\beta$.

The objective function deciding the clean cycle is based on the average
net margin (Mn) over the total time length of the operation period
including the cleaning cycle (t<sub>1</sub>) and cleaning time
(t<sub>2</sub>). That is,

$$Mn\  = \ \frac{M_{G}\  - \ E\  - \ C}{t_{1} + t_{2}}$$

where M<sub>G</sub> is the gross margin in dollars, and E and C are the
costs of energy and exchanger cleaning, respectively.

This can be expressed in an integral form as

$$M_{n} = \frac{\int_{0}^{t1}{\{ m1\  - \ E(t)\} dt\  - \ \sum_{i}^{}C_{L,i}}}{t_{1} + t_{2}} = \frac{\int_{0}^{t1}{\{ m1\  - \ \lbrack A\  - \ B\ .T^{N}(t)\rbrack\} dt\  - \ \sum_{i}^{}C_{L,i}}}{t_{1} + t_{2}}$$

where m<sub>1</sub> is the average gross margin in dollars per day. C<sub>L,i</sub> is the
constant cleaning

cost for exchanger i.

Let K=$\sum_{i}^{}C_{L,i}$ and T<sup>N</sup>(t) be a linear function of
$T_{0}^{N}$ , the integral of the previous equation yields

$$M_{n} = \frac{(m_{1}\  - \ \lbrack A\  - \ B\ .T^{N}(t))*t_{1}\  - \ 0.5\lambda B\ t_{1}^{2} - K}{t_{1} + t_{2}}$$

Therefore, the optimal clean cycle t<sub>1</sub> can be determined from the
derivative of M<sub>n</sub>

over t1:

$$\frac{dM_{n}}{dt_{1}}\  = \ 0$$

which leads to

$$0.5\lambda B\ t_{1}^{2}\  + \ (0.5\lambda Bt_{2})\ t_{1}\  + \ \lbrack - K - (m_{1} - \ A\ \  + \ BN_{0})t_{2}\  = \ 0$$

Solving previous equation for t1 gives the optimal clean cycle denoted as
${\widehat{t}}_{1}$. (The positive solution ofcourse)

The average energy cost (k\$/day) can be found from

$$\overline{E} = \frac{\int_{0}^{t1}{\lbrack A\  - \ B\ .T^{N}(t)\rbrack\} dt}}{t_{1} + t_{2}}\  = \ \frac{(A\  - \ B\ .T_{0}^{N})\ {\widehat{t}}_{1}\  + 0.5\lambda B{{\widehat{t}}_{1}}^{2}\ \ }{{\widehat{t}}_{1} + \ t_{2}}$$

The network terminal temperature at ${\widehat{t}}_{1}$. just before
cleaning is

$$T^{N}({\widehat{t}}_{1}) = T_{0}^{N}\  - \ \lambda.{\widehat{t}}_{1}$$

Sample calculations from the method: The model is implemented in the
spreadsheet

(Table 7.1) in which various cases can be studied for the purpose of
illustration. Note the numbers in bold are manual input, while the rest
are calculated based on the model.
![Example Image](imgs/heat_train1_table.png) <br>
In this example, two hypothetical cases are compared. The value of lambda in
the first case of 0.2 \[<sup>o</sup>C/day\] is on the high side, but it
is not unrealistic as values as high as this one have been observed in
processing very heavy feed in the crude unit in a refining plant. The
spreadsheet finds an optimal cleaning cycle of 121 days. Given the
maximum furnace duty of 160 MW, one finds that the cleaning after 121
days is justified purely on energy costs!

In the second case, the value of λ is reduced to 0.15
\[<sup>o</sup>C/day\] because heavy fouling exchangers were replaced
with more advanced heat exchangers. At a lower fouling rate λ, three
effects can be observed:

-   The optimum run length found from the model is longer (141 vs. 121
    days).

-   T<sup>N</sup> at ${\widehat{t}}_{1}$ is higher than that of case 1
    (259 <sup>o</sup>C vs. 256 <sup>o</sup>C).

-   Lower λ (slower rate in fouling) yields both energy and cleaning
    cost savings.

```python
import pyomo.environ as pyo
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
m_1 = 205
A = 118.6
B = 0.3
T_N0 = 280
lambda_decay = 0.2
K = 100
t_2 = 2
t1_array = np.arange(300)
M_n = (((m_1-A+B*T_N0)*t1_array)-(0.5*lambda_decay*B*(t1_array**2)) - K) / (t1_array+t_2)

plt.plot(t1_array,M_n);
```

```python


model = pyo.ConcreteModel()
model.t_1 = pyo.Var(within=pyo.NonNegativeReals)
m_1 = 205                       # c/day
A = 118.6                       # k$/day
B = 0.3                         # k$/day/C
T_N0 = 280                      # C
lambda_decay = [0.2,0.15]       # C/day
K = 100                         # k$
t_2 = 2                         # days
Q_N0 = 160                      # MW
Beta = 0.1
t1_opt = []
T_N = []
E = []
Q_N = []
M_n_opt = []
for i in lambda_decay:
    M_n = (((m_1-A+B*T_N0)*model.t_1)-(0.5*i*B*(model.t_1**2)) - K) / (model.t_1+t_2)
    model.obj = pyo.Objective(expr = M_n , sense=pyo.maximize)
    status = pyo.SolverFactory('ipopt').solve(model)
    pyo.assert_optimal_termination(status)
    t1_opt.append(model.t_1.value)
    T_N.append(T_N0 - i*model.t_1.value)
    E.append(((A-B*T_N0)*model.t_1.value + 0.5*i*B*model.t_1.value**2)/(model.t_1.value+t_2))
    Q_N.append(Q_N0 - Beta*model.t_1.value)
    M_n_opt.append(M_n()*365/1000)


table_data = [
    ["lambda", "Decay of TN", "0.2",lambda_decay[1], "0.15",lambda_decay[1], "°C/day"],
    ["TN 0", "Clean terminal temperature", T_N0 , T_N0 , T_N0 , T_N0 , "°C"],
    ["t_2", "Down time for cleaning",t_2,t_2,t_2,t_2, "days"],
    ["K", "Cost of cleaning (not margin loss)", K, K, K, K,"k$"],
    ["m1", "Gross margin during normal operation", m_1, m_1,m_1, m_1, "k$/day"],
    ["m2", "Gross margin during cleaning", "0.0", "0.00","0.0", "0.00", "k$/day"],
    ["A", "Energy cost constant", A, A,A, A, "k$/day"],
    ["B", "Energy cost constant", B, B,B, B, "k$/day/°C"],
    ["t_1_opt", "Optimal clean cycle", "121.0",t1_opt[0], "140.00",t1_opt[1], "days"],
    ["TN (t_1_opt)", "Optimal network terminal temperature", "256.0",T_N[0], "259.00",T_N[1], "°C"],
    ["Q_opt", "Furnace duty at in", "149.0",Q_N[0], "146.00",Q_N[1], "MW"],
    ["E", "Average energy costs", "40.9",E[0], "40.60",E[1], "k$/day"],
    ["M", "Yearly net margin", "58.5",M_n_opt[0], "58.80", M_n_opt[1],"M$/year"]
]


df = pd.DataFrame(table_data, columns=["Symbol", "Description", "Book (Case 1)", "Calculated (Case 1)", "Book (Case 2)", "Calculated (Case 2)", "Units"])
df
    
```


**REVISED MODEL FOR CLEAN CYCLE OPTIMIZATION**

The model described in the previous section assumes that the plant will
shut down during cleaning of the exchanger train. This is usually not
the case in reality where only some of the heat exchangers are taken out
for cleaning, while the process unit is kept running at reduced
throughput. Figure 7.4 is considered where there are two parallel trains
of heat exchangers. It is assumed that during cleaning, one train is
kept in operation while the other train of exchangers go through
cleaning in rotation. To minimize throughput loss the bypass lines are
opened slightly to maintain the terminal pressure above the bubble point
pressure to avoid feed flashing before furnace. Minimum amount of vapor
(\<3%) before furnace is allowed in order for control of furnace flow
paths. The bypass is opened until the furnace hits the maximum duty.<br>
![Example Image](imgs/heat_train2.png) <br>
In the case that one train is cleaned while the other train continues to
operate, the net margin in period t2 is no longer zero and therefore an
extra term is introduced to the M<sub>n</sub> equation to yield

$$M_{n} = \frac{\int_{0}^{t1}{\{ m_{1}\  - \ \lbrack A\  - \ B\ .T^{N}(t)\rbrack\} dt\  + \ \int_{0}^{t2}{\lbrack m_{2}\  - \ E_{2}(t)\rbrack.dt} - \ K}}{t_{1} + t_{2}}$$

where E<sub>2</sub> is the energy cost during the cleaning, which is a
constant because we know that the furnace will be running at maximum
duty during cleaning time t<sub>2</sub>. The gross margin m<sub>2</sub>
during the cleaning will be lower than m1 since the throughput will be
lower. How far the throughput has to be reduced will depend on the
actual situation, and will have to be estimated from historical data.
For simplicity, m2 is considered as a constant since t<sub>2</sub> is
much shorter in time compared to t<sub>1</sub>.

The integral of equation yields

$$M_{n} = \frac{(m_{1}\  - \ \lbrack A\  - \ B\ .T^{N}(t))*t_{1}\  - \ 0.5\lambda B\ t_{1}^{2}\  + \ (m_{2}\  - \ E_{2}){*t}_{2}\  - K}{t_{1} + t_{2}}$$

A pseudo-cost of cleaning, K”, is introduced as

$$K"\  = \ K\  - \ (m_{2}\  - \ E_{2}){*t}_{2}$$

This K” can be negative when the cleaning cost is compensated by the
unit margin made from the unit running on one preheat train.

The average energy cost at the optimal cleaning time is calculated as

$$\overline{E} = \frac{\int_{0}^{t1}{\lbrack A\  - \ B\ .T^{N}(t)\rbrack\}\ .\ \ dt}\  + \ \int_{0}^{t2}{E_{2}(t).dt}}{t_{1} + t_{2}}\  = \ \frac{(A\  - \ B\ .T_{0}^{N})\ {\widehat{t}}_{1}\  + 0.5\lambda B{{\widehat{t}}_{1}}^{2}\  + E_{2}*t_{2}}{{\widehat{t}}_{1} + \ t_{2}}$$

Sample calculation from revised model: The change can be made readily to
the spreadsheet, which is shown in Table 7.2.<br>
![Example Image](imgs/heat_train2_table.png) <br>
The input values are kept the same as in Table 7.1 to allow a fair
comparison between the two models. The difference of the yearly net
margin for these two models indicates that it makes sense to clean one
train while keeping the other train running instead of shutting down the
unit. Furthermore, the optimal cleaning cycle for the revised model is
much shorter (t<sub>1</sub> = 87 days for λ = 0.2). Interestingly, this
cleaning cycle is way lower than the current practice for crude unit
preheat train, but well in line with practice in the early 1990s. Again
the model justifies cleaning solely justified on energy loss.

As a remark, the above model can be adjusted to reflect different
situations:

1.  where only the most fouled exchangers are selected to clean;

2.  where time variant margin is estimated instead of average margin;
    and

3.  Other significant costs are introduced.

```python
import pyomo.environ as pyo
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
m_1 = 205                       # k$/day
m_2 = 151                       # k$/day
A = 118.6                       # k$/day
B = 0.3                         # k$/day/C
T_N0 = 280                      # C
lambda_decay = [0.2,0.15]       # C/day
K = 100                         # k$
t_2 = 2                         # days
E_2 = 38                        # k$/day
Q_N0 = 160                      # MW
Beta = 0.1
t1_array = np.arange(400)
M_n = M_n2 = ((m_1-A+B*T_N0)*t1_array-0.5*i*B*(t1_array**2)+(m_2-E_2)*t_2 - K) / (t1_array+t_2)

plt.plot(t1_array,M_n);
```

```python

import pyomo.environ as pyo

model2 = pyo.ConcreteModel()
model2.t_1 = pyo.Var(within=pyo.NonNegativeReals)
m_1 = 205                       # k$/day
m_2 = 151                       # k$/day
A = 118.6                       # k$/day
B = 0.3                         # k$/day/C
T_N0 = 280                      # C
lambda_decay = [0.2,0.15]       # C/day
K = 100                         # k$
t_2 = 2                         # days
E_2 = 38                        # k$/day
Q_N0 = 160                      # MW
Beta = 0.1
t1_opt = []
T_N = []
E = []
Q_N = []
M_n_opt = []
for i in lambda_decay:
    M_n2 = ((m_1-A+B*T_N0)*model2.t_1-0.5*i*B*(model2.t_1**2)+(m_2-E_2)*t_2 - K) / (model2.t_1+t_2)
    model2.obj = pyo.Objective(expr = M_n2 , sense=pyo.maximize)
    status = pyo.SolverFactory('ipopt').solve(model2)
    pyo.assert_optimal_termination(status)
    t1_opt.append(model2.t_1.value)
    T_N.append(T_N0 - i*model2.t_1.value)
    E.append(((A-B*T_N0)*model2.t_1.value + 0.5*i*B*model2.t_1.value**2)/(model2.t_1.value+t_2))
    Q_N.append(Q_N0 - Beta*model2.t_1.value)
    M_n_opt.append(M_n2()*365/1000)


table_data = [
    ["A", "Decay of TN", "0.2",lambda_decay[1], "0.15",lambda_decay[1], "°C/day"],
    ["TN 0", "Clean terminal temperature", T_N0 , T_N0 , T_N0 , T_N0 , "°C"],
    ["t_2", "Down time for cleaning",t_2,t_2,t_2,t_2, "days"],
    ["K", "Cost of cleaning (not margin loss)", K, K, K, K,"k$"],
    ["m1", "Gross margin during normal operation", m_1, m_1,m_1, m_1, "k$/day"],
    ["m2", "Gross margin during cleaning", m_2, m_2,m_2,m_2, "k$/day"],
    ["A", "Energy cost constant", A, A,A, A, "k$/day"],
    ["B", "Energy cost constant", B, B,B, B, "k$/day/°C"],
    ["t_1_opt", "Optimal clean cycle", "87",t1_opt[0], "101",t1_opt[1], "days"],
    ["TN (71)", "Optimal network terminal temperature", "263",T_N[0], "265",T_N[1], "°C"],
    ["Q_opt", "Furnace duty at in", "143",Q_N[0], "141",Q_N[1], "MW"],
    ["E", "Average energy costs", "40.8",E[0], "40.5",E[1], "k$/day"],
    ["M", "Yearly net margin", "59.2",M_n_opt[0], "59.4", M_n_opt[1],"M$/year"]
]

df = pd.DataFrame(table_data, columns=["Symbol", "Description", "Book (Case 1)", "Calculated (Case 1)", "Book (Case 2)", "Calculated (Case 2)", "Units"])
df

```
